# Encoding categorical features

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
train_set['warehouse'] = le.fit_transform(train_set['warehouse'])
test_set['warehouse'] = le.transform(test_set['warehouse'])
```

```python
train_set.head()
```

```python
test_set.head()
```

```python
def preprocessing(data):
    
    data['date'] = pd.to_datetime(data['date'])
    data['day'] = data['date'].dt.day
    data['month'] = data['date'].dt.month
    data['year'] = data['date'].dt.year
    
    data['day_of_week'] = data['date'].dt.dayofweek
    data['week_of_year'] = data['date'].dt.isocalendar().week
    data['is_weekend'] = data['day_of_week'].isin([5, 6]).astype(int)
    data['is_holiday'] = data['holiday'].astype(int)
    data['quarter'] = data['date'].dt.quarter
    
    return data
```

```python
train_set = preprocessing(train_set)
test_set = preprocessing(test_set)
```
